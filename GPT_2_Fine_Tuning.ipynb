 Vamos treinar o nosso modelo! Neste caso, usaremos o conjunto de dados de atendimento ao consumidor. Para alterar o conjunto de dados em que os modelos GPT-2 serão ajustados, você pode substituir esta URL por outro arquivo .txt e modificar a parte correspondente na célula seguinte. Lembre-se de que você pode usar conjuntos de dados pequenos, mas deve ter cuidado para não executar o ajuste fino por muito tempo, pois isso pode resultar em sobreajuste. Em geral, espere obter resultados interessantes em minutos a horas, dependendo do tamanho do conjunto de dados (geralmente entre 1-10 megabytes). Se você estiver usando um conjunto de dados menor, pode ser necessário interromper o ajuste fino mais cedo, pois ele pode ser muito rápido.

```
from datasets import load_dataset

dataset_name = "Kaludi/Customer-Support-Responses"

dataset = load_dataset(dataset_name)

train_dataset = dataset['train']

formatted_data = []

for example in train_dataset:
    query = example['query']
    response = example['response']

    formatted_example = f"{query} <SOS> {response} <EOS>"

    formatted_data.append(formatted_example)

formatted_text = "\n".join(formatted_data)

with open("dataset.txt", "w") as file:
    file.write(formatted_text)
```

```
from googletrans import Translator

# Função para traduzir um texto para o português
def translate_to_portuguese(text):
    translator = Translator()
    translated = translator.translate(text, src='en', dest='pt')
    return translated.text

# Ler o conjunto de dados original
with open('dataset.txt', 'r', encoding='utf-8') as file:
    dataset = file.read()

# Dividir o conjunto de dados em linhas (se estiver em formato de linhas)
lines = dataset.split('\n')

# Traduzir cada linha para o português e armazenar em uma lista
translated_lines = [translate_to_portuguese(line) for line in lines]

# Juntar as linhas traduzidas de volta em um único texto
translated_dataset = '\n'.join(translated_lines)

# Salvar o conjunto de dados traduzido em um arquivo
with open('dataset-traduzido.txt', 'w', encoding='utf-8') as file:
    file.write(translated_dataset)
```
